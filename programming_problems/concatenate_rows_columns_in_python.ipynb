```python
import numpy as np
import pandas as pd
```

```python
# initialize a dataframe
df = pd.DataFrame({'a':[0, 1, 2, 3], 'b':[0, 10, 20, 30]})
df
```

```python
# initialize a series named "c"
ts = pd.Series(data=['A', 'B', 'C', 'D', 'E'], index=[0, 2, 4, 6, 8], name='c')
ts
```

```python
# concatenate ts and df and call the new dataframe df2
df2 = pd.concat(objs=[ts, df], axis=1)
df2
```

```python
# add another column called "id" to df2
df2['id'] = range(1,len(df2)+1)
df2
```

```python
# we want to use column "id" as our index
df2 = df2.set_index(keys='id')
df2
```

```python
# we have another series called "d"
ts2 = pd.Series(data=['a', 'b', 'c', 'd', 'e'], index=[5, 7, 9, 11, 13], name='d')
ts2
```

```python
# concatenate df2 with ts2, joining by index
pd.concat(objs=[df2, ts2], axis=1)
```

```python
# concatenate df2 with ts2, joining by index. 
# Note that this will give an error message because the index is not unique!!!
pd.concat(objs=[df2, ts2], axis=0)
```

```python
# A better way to do it is to "ignore" the index.
# In this case, the index of the new dataframe will be reset.
# However, it discard the series' name "c" and show 0 instead.
pd.concat(objs=[df2, ts2], axis=0, ignore_index=True)
```

```python
# even the series' name matches with the existing column, it will still shows 0.
ts3 = pd.Series(data=['x', 'y', 'z'], index=[5, 7, 9], name='c')
pd.concat(objs=[df2, ts3], axis=0, ignore_index=True)
```

```python
# However, if we convert the series to dataframe, the column "c" will be matched.
# This said, column name of dataframe (columns) and name of series (names) are different thing!!
df_ts3 = pd.DataFrame(data={'c':ts3})
# which is equivalent to the expression below
# df_ts3 = pd.DataFrame(data=['x', 'y', 'z'], index=[5, 7, 9], columns=['c'])

pd.concat(objs=[df2, df_ts3], axis=0, ignore_index=True)
```

```python
# concat would match all the columns and append the new columns not existed yet.
df4 = pd.DataFrame(data={'a':[-1, -2], 'b':[-10, -20], 'd':['m', 'n']})
pd.concat(objs=[df2, df4], axis=0, ignore_index=True)
```

```python
# append np array as a new column to a dataframe
df2['log1'] = (np.arange(7)>3)+0
df2
```

```python
# This is a solid way to add new column
# first make a series with the same index, then concat it
logic2 = (np.arange(7)>5)+0
logic2 = pd.Series(data=logic2, index=df2.index, name='log2')
pd.concat(objs=[df2, logic2], axis=1)
```
